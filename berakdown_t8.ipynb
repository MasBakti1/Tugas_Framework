```python
import pandas as pd

file_path = r"C:\Users\user x\Desktop\Pemdas\Tugas_Framework\disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx"
df = pd.read_excel(file_path)

print(df)
```

```python
df['jumlah_produksi_sampah'] = pd.to_numeric(df['jumlah_produksi_sampah'], errors='coerce')
Dsampah = pd.DataFrame(
    {
        'Lokasi': df['nama_kabupaten_kota'],
        'Sampah (Ton/Hari)': df['jumlah_produksi_sampah'],
        'Tahun': df['tahun']
    }
)

print(Dsampah)
```

```python
Dsampah.to_csv('hasil_analisa_total_produksi_sampah.csv', index=False)
Dsampah.to_excel('hasil_analisa_total_produksi_sampah.xlsx', index=False)

total_per_tahun_df = pd.DataFrame(list(total_per_tahun.items()), columns=['Tahun', 'Total Sampah (Ton/Hari)'])
total_per_tahun_df.to_csv('total_per_tahun.csv', index=False)

total_per_kabupaten_per_tahun_df = pd.DataFrame(list(kabupaten_per_tahun.items()), columns=['Lokasi-Tahun', 'Total Sampah (Ton/Hari)'])
total_per_kabupaten_per_tahun_df.to_csv('kabupaten_per_tahun.csv', index=False)
```
